```python
from sklearn.model_selection import train_test_split
X = tele_users2[['gender', 'SeniorCitizen', 'Partner',
       'Dependents', 'tenure', 'PhoneService', 'MultipleLines',
       'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
       'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
       'PaperlessBilling', 'PaymentMethod', 'MonthlyCharges']] 
y = tele_users2['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=654)
```

F.

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.fit(X_train,y_train)
```

```python
para_test = {
    'n_estimators': [50, 70, 100],
    'max_depth': [4, 6, 8],
    'max_features': [12, 14, 16], # max features should be lower than 17
    'min_samples_leaf': [6, 10, 14]
}
```

```python
from sklearn.model_selection import GridSearchCV
CV_rfc = GridSearchCV(estimator=clf, param_grid=para_test, cv= 5)
CV_rfc.fit(X_train, y_train)
print(CV_rfc.best_params_)
```

```python
CV_rfc.best_score_
```

```python
Rainf_clf=RandomForestClassifier(n_estimators=50, max_depth=6, max_features=14, min_samples_leaf=10, random_state=654)
Rainf_clf.fit(X_train,y_train)
```

O:

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(random_state=654)
logmodel.fit(X_train2, y_train2)
```

P:

```python
logmodel.coef_
```

a:Tenure has a slight negative impact on Churn. The more tenure years they have, they are less likely to churn. MonthlyCharges has a minor positive influence on the Churn.The more expense they pay each month, the more they intend to churn.

b:Contract has a strongly negative impact on Churn, So the longer contract they signed, the less likely they churn. PaymentMethod has an slightly positive impact on churn. OnlineSecurity has an little negative impact on churn,the more they care about security, the less they are likely to churn. So does TechSupport, the more they need tech support, the less they intend to churn. As for OnlineBackup, things are similar, the more they need online backup, the less they are likely to churn. Gender has a minor positive impact on churn,which means female are less likely to churn than male. Women are more loyal. And DeviceProtection has a minor positive on churn as well. It means the more they need device protection, the more likely they intend to churn. MultipleLines shows a positive impact on churn,which means, the less automatic(or outdated) the paymethod they choose, they less they want to churn, PaperlessBilling shows a positive impact on churn. The more they like paperless billing, the more likely they churn.

```python
prediction_log = logmodel.predict(X_test2)
matrix_log = confusion_matrix(y_test2, prediction_log)
sns.heatmap(matrix_log, fmt='g', square=True, annot=True, cbar=False)
plt.xlabel("Predicted Result")
plt.ylabel("Actual Result")
a, b = plt.ylim() 
a += 0.5 
b -= 0.5 
plt.ylim(a, b)
plt.show()
```

```python
print("Accuracy:",metrics.accuracy_score(y_test2, prediction_log))
print("Precision:",metrics.precision_score(y_test2, prediction_log))
print("balanced accuracy:",metrics.balanced_accuracy_score(y_test2, prediction_log))
print("Sensitivity:",metrics.recall_score(y_test2, prediction_log))
```

```python
tn, fp, fn, tp = confusion_matrix(prediction_log, y_test2).ravel()
print("Specificity: {:.4f}".format(tn/(tn+fp)))
```
